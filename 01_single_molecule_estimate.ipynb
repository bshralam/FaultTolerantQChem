```python
import pennylane as qml
import numpy as np

symbols = ['O', 'H', 'H']
geometry = np.array([[0.00000000,  0.00000000,  0.28377432],
                     [0.00000000,  1.45278171, -1.00662237],
                     [0.00000000, -1.45278171, -1.00662237]])

mol = qml.qchem.Molecule(symbols, geometry, basis_name='sto-3g')
core, one, two = qml.qchem.electron_integrals(mol)()

algo = qml.estimator.DoubleFactorization(one, two)
print("lambda:", algo.lamb)
print("Toffoli:", algo.gates)
print("qubits:", algo.qubits)
```

```python
for err in [0.01, 0.001, 0.0001]:
    a = qml.estimator.DoubleFactorization(one, two, error=err)
    print(f"error={err}: Toffoli={a.gates:.3e}, qubits={a.qubits}")
```

```python
mol2 = qml.qchem.Molecule(symbols, geometry, basis_name='6-31g')
core2, one2, two2 = qml.qchem.electron_integrals(mol2)()
algo2 = qml.estimator.DoubleFactorization(one2, two2)
print("STO-3G: qubits=%d, Toffoli=%.3e" % (algo.qubits, algo.gates))
print("6-31G:  qubits=%d, Toffoli=%.3e" % (algo2.qubits, algo2.gates))
```
